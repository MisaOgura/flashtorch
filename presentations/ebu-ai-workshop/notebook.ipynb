## Explainability & AI

What is it and why does AI need to be explainable?

---

## Explainability

Can you **explain** why a system behaves the way it does?

---

Interpretability: can you _predict_ an outcome, not (necessarily) knowing why?

e.g. Driving a car:
- I _know if_ I steer to the left, _then_ the car turns left == **Interpretability**
- I can _explain_ why that is (i.e. physical mechanics behind it) == **Explainability**

**In the context of AI:**

Ability to explain decisions made by algorithms.

<aside class="notes">
    Same as interpretability?<br/>
    No, it asks a different question.<br/>
    <br/>
    My talk today will focus on explainability.
</aside>

## Saliency maps for CNNs

Gives some intuition of _attention_

---

- First <a target='_blank' href='https://arxiv.org/abs/1312.6034'>Introduced</a> in 2013

- Calculates the _relationship_ (or gradients) between output label(s) & input image

- **Positive gradients == positive effects** on the label of interest

<img id='cnn' src='images/cnn2.png' width='800'>

<div class='source'>
    <a target='_blank' href='https://missinglink.ai/guides/convolutional-neural-networks/convolutional-neural-network-tutorial-basic-advanced'>Source</a>
</div>

<aside class="notes">
    Mathematical property of gradients!<br/>
    Choose a class of interest.<br/>
    Calculate gradients for every pixels.<br/>
    Highlight pixels with positive gradients.<br/>
    You get a saliency map!
</aside>

````python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torchvision.models as models
````

### Prepare an input üê¶

---

<aside class="notes">
    Load & transform the input image.<br/>
    Right type, right shape - ready to be fed into the CNN.
</aside>

````python
from flashtorch.utils import load_image, apply_transforms, denormalize, format_for_plotting

image = load_image('../../examples/images/great_grey_owl.jpg')

owl = apply_transforms(image)

print(f'Before: {type(image)}')
print(f'After: {type(owl)}, {owl.shape}')

plt.imshow(format_for_plotting(denormalize(owl)))
plt.title('Input tensor')
plt.axis('off');
````

### Create a `Backprop` object with a pre-trained model

---

````python
from flashtorch.saliency import Backprop

model = models.alexnet(pretrained=True)

backprop = Backprop(model)
````

- Registers **custom functions** to model layers
- Grabs gradients out of the computational graph

<img id='cnn' src='images/cnn2.png' width='800'>

<aside class="notes">
    PyTorch doesn't expose intermediate gradients.
</aside>

### Visualize the gradients

Our first saliency map!

---

````python
from flashtorch.utils import ImageNetIndex 

imagenet = ImageNetIndex()
target_class = imagenet['great grey owl']

backprop.visualize(owl, target_class, guided=True)
````

### Building a flower classifier

---

1. Get a pre-trained model (ImageNet: 1000 classes)
2. Swap out the last layer (102 classes) --> _Un-tuned_
3. Train with a <a href='http://www.robots.ox.ac.uk/~vgg/data/flowers/102/index.html' target='_blank'>flower dataset</a> --> _Tuned_


Un-tuned model: **0.1%** test accuracy üòÖ

_Why_ is it so bad? Let's take a look.

<aside class="notes">
    With transfer learning, I created a flower classifier.<br/>
    <br/>
    It's easy to jump straight on to training.<br/>
    But let's stop and inspect why.
</aside>

````python
plt.imshow(load_image('images/foxgloves.jpg'))
plt.title('Foxgloves')
plt.axis('off');
````

````python
def create_model(filepath=None):
    in_feats = 2208
    out_feats = 102  # Total num of classes

    if filepath is not None:
        model = models.densenet161()
    else:
        model = models.densenet161(pretrained=True)
        
    model.classifier = nn.Sequential(nn.Linear(in_feats, out_feats),
                           nn.LogSoftmax(dim=1))

    if filepath is not None:
        checkpoint = torch.load(filepath, map_location='cpu')
        model.load_state_dict(checkpoint['state_dict'])
        
    for param in model.parameters():
        param.requires_grad = False
    
    return model
````

### Un-tuned model

Test accuracy: 0.1%

---

````python
image = load_image('images/foxgloves.jpg')

foxglove = apply_transforms(image)

class_index = 96  # foxglove

untuned_model = create_model()
tuned_model = create_model('../../models/flower_classification_transfer_learning.pt')
````

````python
backprop = Backprop(untuned_model)

backprop.visualize(foxglove, class_index, guided=True)
````

<div class='bottom'>
    After training, the tuned model achieved <b>98.7%</b> test accuracy üéâ<br/>
    But <i>how</i>? ü§î
</div>

### Tuned model

Test accuracy: 98.7%

---

````python
backprop = Backprop(tuned_model)

backprop.visualize(foxglove, class_index, guided=True)
````

<div class='bottom'>
    The model has <i>learnt</i> to focus on to the <b>most distinguising pattern</b>.<br/>
    <br/>
    This explains!
</div>
